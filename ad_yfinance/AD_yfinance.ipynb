```python
import pandas as pd
import numpy as np
import yfinance as yf
import math
```

```python
acoes_listadas = pd.read_html("https://www.dadosdemercado.com.br/acoes")
```

```python
df_acoes = acoes_listadas[0]
df_acoes = df_acoes[['Ticker', 'Nome']]
```

```python
def fetch_market_cap(ticker_list):
    data = yf.download(ticker_list, period = '1d', group_by='ticker', auto_adjust=False)
    stocks_data = []

    for ticker in ticker_list:
        latest_price = data[ticker]['Close'].iloc[-1]
        market_cap = yf.Ticker(ticker).info.get("marketCap", 'N/A')
        stocks_data.append({
            "Ticker" : ticker,
            "Market Cap" : market_cap,
            "Latest Price" : latest_price,
        })

    stocks_df = pd.DataFrame(stocks_data)

    return stocks_df
```

```python
tickers_list = [ticker + ".SA" for ticker in df_acoes["Ticker"].values.tolist()[:-350]]
df = fetch_market_cap(tickers_list)
```

```python
df = df.sort_values(by = "Market Cap", ascending = False)
```

```python
df = df.head(10)
df.reset_index(inplace=True, drop = True)
```

```python
portfolio_size = int(input("Enter the amount you want to invest: "))
position_size =  portfolio_size / len(df.index)
```

```python
df.loc[:, 'Number of Shares to buy'] = df['Latest Price'].apply(lambda price: math.floor(position_size / price))
```
